```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=[10,20,30,40]
y=["c++","c","python","Java"]
```

```python
ex=[0.4,0.0,0.0,0.0]
c=["r","b","g","y"]
plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.1f %%",shadow=True,radius=1.2,labeldistance=1.2,
        textprops={"fontsize":15})
plt.show()
```
